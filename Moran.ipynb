```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def frange(start,final,interval):
    
    numbers = []
    while start < final:
        numbers.append(start)
        start = start + interval
        
    return numbers

def moran(populationSize,steps,increments) :
    groupA = np.floor(populationSize/2)
    groupB = populationSize - groupA
    groupASize = [groupA]
    for i in range(steps):
        if random.uniform(0,1) < groupA / (groupA + groupB) :
            groupA += 1
        else:
            groupB += 1
        if random.uniform(0,1) < groupA / (groupA + groupB) :
            groupA -= 1
        else: 
            groupB -= 1
        if i % increments == 0 :
            groupASize.append(groupA)
            
    return groupASize
```

```python
result = moran(200,25000,100)
print(result)
x = frange(0,25001,100)
plt.plot(x,result)
plt.show()
```

```python
def moran_experiment(populationSize,steps,repetitions,increments):
    filename = input('Enter a name for the file: ')
    results = []
    for i in range(repetitions):
        results.append(moran(populationSize,steps,increments))
    x = frange(0,steps+1,increments)
    for i in range(len(results)):
        plt.plot(x,results[i])
    plt.ylim(0,populationSize)
    plt.savefig(filename)
    plt.show()
```

```python
moran_experiment(2000,7000000,15,100)
```

```python
moran_experiment(10,1000,15,1)
```
